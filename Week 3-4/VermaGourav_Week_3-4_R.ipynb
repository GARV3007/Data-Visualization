Line and Step chart using R


```r
# Import libraries
library(ggplot2)
library(readxl)
```

```r
# Read the data
population <- read_excel('world-population.xlsm')
head(population)
```

```r
# Line Chart
ggplot(data=population, aes(x=Year, y=Population, group=1))+
geom_line(color="red")+
geom_point()
```

```r
# Step chart
ggplot(data=population, aes(x=Year, y=Population, group=1))+
geom_step(color="red")
```
